# Different Stats Configuration for metrics

The following document illustrates the different configurations of the map_control_metric and the bomb distance metric.

## (Preparation, loading data)

```python
import os
from models.data_manager import DataManager
demo_path = os.path.join(os.getcwd(), '../demos/esta/lan/de_dust2/00e7fec9-cee0-430f-80f4-6b50443ceacd.json')
dm = DataManager(demo_path, do_validate=False)
```

## Map Control Metric

```python
from metrics.map_control_metric import MapControlMetric
mpc = MapControlMetric()

# Default  Configuration:
mpc.process_metric_frame(dm, 3, 10, True)
```

```python
from metrics.map_control_metric import MapControlMetric
mpc = MapControlMetric()

# With different area treshholds (maximum area of map that can be controlled
mpc.process_metric_frame(dm, 3, 10, True, 1/15, 10)
mpc.process_metric_frame(dm, 3, 10, True, 1/10, 10)
mpc.process_metric_frame(dm, 3, 10, True, 1/5, 10)
```

```python
from metrics.map_control_metric import MapControlMetric
mpc = MapControlMetric()

# With different round normalization parameter (normalization and occupied only)
mpc.process_metric_round(dm, 3, True, 1/20, 10, norm=0)
mpc.process_metric_round(dm, 3, True, 1/20, 10, norm=1)
mpc.process_metric_round(dm, 3, True, 1/20, 10, norm=2)

```

```python
from metrics.map_control_metric import MapControlMetric

# With different round occupied parameter and absolute
mpc = MapControlMetric()
mpc.process_metric_round(dm, 3, True, 1/20, 10, occupied_only=False, norm=2, absolute=True)

mpc = MapControlMetric()
mpc.process_metric_round(dm, 3, True, 1/20, 10, occupied_only=True, norm=2, absolute=True)
```
